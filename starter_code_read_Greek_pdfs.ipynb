### Διάβασμα των pdf και δημιουργεία DataFrame

````python
list_of_results = []
for pdf in list_of_files:
    result = {}
    results = parser.from_file(pdf)
    metadata = results.get('metadata')
    result['status'] = results['status']
    result['content'] = results.get('content')
    result['filename'] = os.path.basename(pdf)
    list_of_results.append(result)

df = pd.DataFrame(list_of_results)
````

### Regex

````python
import re

# Regex για την αναγνώριση του μοναδικού κωδικού
pattern = r"Μοναδικός Κωδικός: (\w+)"

'''
To (\w+) εξάγει έναν ή περισσότερους χαρακτήρες (αλφαριθμητικούς και υπογράμμιση) μετά το "Μοναδικός Κωδικός:".
Ότι μπαίνει μέσα σε παρενθέσεις, ουσιαστικά αποτελούν τα δεδομένα από το κείμενο που θέλουμε να εξάγουμε. 
Μπορούμε να έχουμε περισσότερες της μίας παρενθέσεις. Τότε κάθε παρένθεση μπορεί να θεωρηθεί ένα group δεδομένων
που θέλουμε να εξάγουμε. Σε αυτή την περίπτωση θα είχαμε πρόσβαση σε κάθε γκρουπ γράφοντας match.group(1), match.group(2) κόκ.
'''

# Αναζήτηση του μοτίβου στο κείμενο
match = re.search(pattern, df['content'][0])

if match:
    unique_code = match.group(1)
    print("Μοναδικός κωδικός:", unique_code)
else:
    print("Δεν βρέθηκε εγγραφή")
````

````python
# Regex με δύο γρουπ
pattern = r"Ημερομηνία Δημοσίευσης Πλειστηριασμού: (\d{2}/\d{2}/\d{4} \d{2}:\d{2}:\d{2})\s+Μοναδικός Κωδικός: (\w+)"
'''
Το \s+ σημαίνει ένα ή περισσότερα κενά στο κείμενο.
Το \d{2} σημαίνει δύο αριθμούς. 
'''
# Αναζήτηση στο κείμενο
match = re.search(pattern, df['content'][0])

if match:
    date_time = match.group(1)
    unique_code = match.group(2)
    print("Ημερομηνία:", date_time)
    print("Μοναδικός κωδικός:", unique_code)
else:
    print("Δεν βρέθηκε εγγραφή.")
````

### Γλωσσάρι Regex με Παραδείγματα

| Χαρακτήρας | Περιγραφή | Παράδειγμα |
|------------|------------|------------|
| `.`        | Αντιπροσωπεύει οποιονδήποτε χαρακτήρα εκτός από νέο-γραμμή. | `α.β` θα ταιριάζει με `αββ`, `αμβ`, `α1β`, κλπ. |
| `^`        | Αντιστοιχεί στην αρχή μιας γραμμής ή συμβολοσειράς. | `^Καλημέρα` θα ταιριάζει με το `Καλημέρα` μόνο αν είναι στην αρχή της γραμμής. |
| `$`        | Αντιστοιχεί στο τέλος μιας γραμμής ή συμβολοσειράς. | `Καληνύχτα$` θα ταιριάζει με το `Καληνύχτα` μόνο αν είναι στο τέλος της γραμμής. |
| `*`        | Αντιστοιχεί στον προηγούμενο χαρακτήρα 0 ή περισσότερες φορές. | `Αβ*γ` θα ταιριάζει με `Αγ`, `Αβγ`, `Αββγ`, κλπ. |
| `+`        | Αντιστοιχεί στον προηγούμενο χαρακτήρα 1 ή περισσότερες φορές. | `Αβ+γ` θα ταιριάζει με `Αβγ`, `Αββγ`, κλπ., αλλά όχι με `Αγ`. |
| `?`        | Αντιστοιχεί στον προηγούμενο χαρακτήρα 0 ή 1 φορά. | `Αβ?γ` θα ταιριάζει με `Αγ` ή `Αβγ`. |
| `{n}`      | Αντιστοιχεί στον προηγούμενο χαρακτήρα ακριβώς n φορές. | `Αβ{2}γ` θα ταιριάζει με `Αββγ`, αλλά όχι με `Αβγ` ή `Αβββγ`. |
| `{n,}`     | Αντιστοιχεί στον προηγούμενο χαρακτήρα τουλάχιστον n φορές. | `Αβ{2,}γ` θα ταιριάζει με `Αββγ`, `Αβββγ`, κλπ. |
| `{n,m}`    | Αντιστοιχεί στον προηγούμενο χαρακτήρα τουλάχιστον n και το πολύ m φορές. | `Αβ{2,3}γ` θα ταιριάζει με `Αββγ` και `Αβββγ`, αλλά όχι με `Αβγ` ή `Αββββγ`. |
| `[...]`    | Αντιστοιχεί σε οποιονδήποτε έναν χαρακτήρα μέσα στις αγκύλες. | `[ΑΒΓ]` θα ταιριάζει με `Α`, `Β`, ή `Γ`. |
| `[^...]`   | Αντιστοιχεί σε οποιονδήποτε έναν χαρακτήρα που δεν είναι μέσα στις αγκύλες. | `[^ΑΒΓ]` θα ταιριάζει με οποιονδήποτε χαρακτήρα εκτός από `Α`, `Β`, ή `Γ`. |
| `\`        | Χρησιμοποιείται για να διαφύγει ειδικούς χαρακτήρες. | `\.` θα ταιριάζει με την τελεία `.`, αντί για οποιονδήποτε χαρακτήρα. |
| `|`        | Λειτουργεί ως λογικό OR ανάμεσα σε δύο μοτίβα. | `Καλημέρα|Καληνύχτα` θα ταιριάζει με `Καλημέρα` ή `Καληνύχτα`. |
| `()`       | Χρησιμοποιείται για ομαδοποίηση χαρακτήρων ή υποεκφράσεων. | `(Αβ)+` θα ταιριάζει με `Αβ`, `ΑβΑβ`, `ΑβΑβΑβ`, κλπ. |
| `\d`       | Αντιστοιχεί σε οποιοδήποτε ψηφίο. | `\d` θα ταιριάζει με `0`, `1`, `2`, κλπ. |
| `\D`       | Αντιστοιχεί σε οποιοδήποτε μη-ψηφίο. | `\D` θα ταιριάζει με οποιονδήποτε χαρακτήρα εκτός από τα ψηφία. |
| `\w`       | Αντιστοιχεί σε οποιονδήποτε αλφαριθμητικό χαρακτήρα και κάτω παύλα. | `\w` θα ταιριάζει με `α`, `β`, `1`, `_`, κλπ. |
| `\W`       | Αντιστοιχεί σε οποιονδήποτε μη-αλφαριθμητικό χαρακτήρα. | `\W` θα ταιριάζει με οποιονδήποτε χαρακτήρα εκτός από τα αλφαριθμητικά και την κάτω παύλα. |
| `\s`       | Αντιστοιχεί σε οποιονδήποτε χαρακτήρα λευκού διαστήματος. | `\s` θα ταιριάζει με κενό, tab, νέο-γραμμή, κλπ. |
| `\S`       | Αντιστοιχεί σε οποιονδήποτε μη-λευκό χαρακτήρα. | `\S` θα ταιριάζει με οποιονδήποτε χαρακτήρα εκτός από τους χαρακτήρες λευκού διαστήματος. |
