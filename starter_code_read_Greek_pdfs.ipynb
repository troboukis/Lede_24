### Διάβασμα όλων των αρχείων pdf και δημιουργεία pd.Series

````python
import pandas as pd
````

````python
list_of_content = []
for pdf in list_of_files:
    parsed = parser.from_file(pdf)
    list_of_content.append(parsed['content'])
df = pd.Series(list_of_content)    
````

````python
df
````

````python
df = df.replace({'\n':' '}, regex=True).replace({'\t':' '}, regex=True).copy()
df
````

### Διάβασμα των pdf και δημιουργεία DataFrame

````python
list_of_results = []
for pdf in list_of_files:
    result = {}
    results = parser.from_file(pdf)
    metadata = results.get('metadata')
    result['status'] = results['status']
    result['content'] = results.get('content')
    result['filename'] = os.path.basename(pdf)
    list_of_results.append(result)

df = pd.DataFrame(list_of_results)
````

````python
df['content'] = df['content'].replace({'\n':' '}, regex=True).replace({'\t':' '}, regex=True)
````
